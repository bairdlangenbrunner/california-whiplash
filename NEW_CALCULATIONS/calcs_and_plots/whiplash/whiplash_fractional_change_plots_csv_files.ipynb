```python
import numpy
import matplotlib.pyplot as mp
import pandas

mp.rcParams.update({'mathtext.default': 'regular'})
% matplotlib inline
```

```python
whiplash_ratios_ncal = pandas.read_csv('csv_files/store_all_whiplash_ratios_dataframe_8020_loop_years_ncal.csv')
whiplash_ratios_ccal = pandas.read_csv('csv_files/store_all_whiplash_ratios_dataframe_8020_loop_years_ccal.csv')
whiplash_ratios_scal = pandas.read_csv('csv_files/store_all_whiplash_ratios_dataframe_8020_loop_years_scal.csv')

whiplash_ratios_bootstrap_ncal = pandas.read_csv('csv_files/bootstrap_whiplash_ratios_8020_pic_ncal.csv')
whiplash_ratios_bootstrap_ccal = pandas.read_csv('csv_files/bootstrap_whiplash_ratios_8020_pic_ccal.csv')
whiplash_ratios_bootstrap_scal = pandas.read_csv('csv_files/bootstrap_whiplash_ratios_8020_pic_scal.csv')
```

```python
bs_ncal = whiplash_ratios_bootstrap_ncal.values
bs_ncal = bs_ncal[:,1:].flatten()
bs_ncal_stdev = numpy.std(bs_ncal, ddof=1)

bs_ccal = whiplash_ratios_bootstrap_ccal.values
bs_ccal = bs_ccal[:,1:].flatten()
bs_ccal_stdev = numpy.std(bs_ccal, ddof=1)

bs_scal = whiplash_ratios_bootstrap_scal.values
bs_scal = bs_scal[:,1:].flatten()
bs_scal_stdev = numpy.std(bs_scal, ddof=1)
```

```python
whiplash_ratios_ncal.describe()
year_labels = list(whiplash_ratios_ncal)[1:]
year_labels_formatted = [i.split('-')[0]+'-\n'+i.split('-')[1] for i in year_labels]
```

```python
fontsize=12
fig = mp.figure(figsize=(6,3))
color='steelblue'

ax = fig.add_subplot(111)
for i in range(1,7):
    ax.scatter([i]*9, whiplash_ratios_ncal.values[:,i], c=color, s=25, zorder=2)

ax.grid()
ax.set_axisbelow(True)
ax.axhline(1, lw=2, c='0.25', zorder=1)

ax.set_ylim(0,2.25)
ax.set_xlim(0,7)
ax.set_yticks([0,0.5,1,1.5,2])
ax.tick_params(labelsize=fontsize)
ax.set_ylabel('whiplash ratio', fontsize=fontsize)
ax.set_xticks([1,2,3,4,5,6])
ax.set_xticklabels(year_labels_formatted)

ax.axhspan(1-2*bs_ncal_stdev, 1+2*bs_ncal_stdev, color='0.75')

ax.text(s='Northern California', x=0.5, y=1.05, ha='center', va='bottom', transform=ax.transAxes, fontsize=fontsize)
fig.tight_layout()
fig.savefig('figs/whiplash_ratios_8020_ncal.pdf', transparent=True, bbox_inches='tight')
```
